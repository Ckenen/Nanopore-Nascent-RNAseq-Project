```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Distribution of mismatch ratio

```python
d = pd.read_csv("../../1_NanoNASCseq/reports/NanoNASCseq_Summary.csv")
d = d[d["Group"] == "MouseBlastocyst"]
d = d[d["UMIs"] >= 5000]
d = d[d["Time"] == 3]
d = d[d["ActD"].isna()]
```

```python
# 0 uM

d1 = d[d["s4U"] == 0]
vs = d1["TC.Ratio"].values
print("Cells:", len(vs))
print("Mean:", np.mean(vs))
print("Std:", np.std(vs))

plt.figure(figsize=(4, 3))
plt.hist(vs, bins=20, edgecolor="black")
plt.xlabel("Mismatch ratio of T-C")
plt.ylabel("Number of cells")
plt.tight_layout()
plt.show()

plt.figure(figsize=(4, 3))
plt.hist(vs, bins=20, range=(0, 0.002), edgecolor="black")
plt.xlabel("Mismatch ratio of T-C")
plt.ylabel("Number of cells")
plt.tight_layout()
plt.show()
```

```python
# 400 uM

d2 = d[d["s4U"] == 400]
vs = d2["TC.Ratio"].values
print("Cells:", len(vs))
print("Mean:", np.mean(vs))
print("Std:", np.std(vs))

plt.figure(figsize=(8, 3))
plt.hist(vs, bins=100, edgecolor="black")
plt.xlabel("Mismatch ratio of T-C")
plt.ylabel("Number of cells")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 3))
plt.hist(vs, bins=100, range=(0, 0.005), edgecolor="black")
plt.xlabel("Mismatch ratio of T-C")
plt.ylabel("Number of cells")
plt.tight_layout()
plt.show()
```
