```python
import os
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import *
import seaborn as sns
```

# Correlation

```python
for name in ["gene", "transcript"]:
    cts = ["ICM-1", "ICM-2", "EPI", "PE", "TE-1", "TE-2", "muralTE", "polarTE"]
    array = []
    for ct in cts:
        m = pd.read_csv("results/halflife/%s_%ss.%s_name.tsv" % (ct, name, name), sep="\t", index_col=0)
        s = m["TPM"]
        s.name = ct
        array.append(s)
    d = pd.concat(array, axis=1).fillna(0)
    d.index.name = "GeneName"
    d = d[["muralTE", "polarTE", "TE-2", "TE-1", "ICM-1", "ICM-2", "PE", "EPI"]]
    corr = d[d.max(axis=1) >= 10].corr()
    print("Min value:", corr.min().min())
    plt.figure(figsize=(5, 4))
    # plt.title("Gene-based")
    sns.heatmap(corr, cmap="bwr", vmin=0.75, vmax=1, square=True, annot=True, fmt=".2f", cbar_kws=dict(label="Correlation"))
    plt.tight_layout()
    plt.savefig("figures/celltype_expression_correlation.%s-based.pdf" % name, dpi=300)
```

# DEGs

## Gene-based

```python
cts = ["TE-1", "TE-2", "muralTE", "polarTE", "ICM-1", "ICM-2", "EPI", "PE"]

for i in range(len(cts) - 1):
    for j in range(i + 1, len(cts)):
        ct1, ct2 = cts[i], cts[j]
        print(ct1, ct2, sep="\t")
        
        m1 = pd.read_csv("results/halflife/%s_genes.gene_name.tsv" % ct1, sep="\t", index_col=0)
        m1["TPM"] = m1["Total"] * 1e6 / m1["Total"].sum()
        m1.columns = ["%s.%s" % (c, ct1) for c in m1.columns]

        m2 = pd.read_csv("results/halflife/%s_genes.gene_name.tsv" % ct2, sep="\t", index_col=0)
        m2["TPM"] = m2["Total"] * 1e6 / m2["Total"].sum()
        m2.columns = ["%s.%s" % (c, ct2) for c in m2.columns]
        
        d = pd.concat([m1, m2], axis=1)
        for x in ["Total", "TPM", "Nascent"]:
            for y in [ct1, ct2]:
                d["%s.%s" % (x, y)] = d["%s.%s" % (x, y)].fillna(0)
        d = d[(d["Total.%s" % ct1] > 0) | (d["Total.%s" % ct2] > 0)]
        d["Mean"] = d[["TPM.%s" % ct1, "TPM.%s" % ct2]].mean(axis=1)
        d["TPM.Max"] = d[["TPM.%s" % ct1, "TPM.%s" % ct2]].max(axis=1)
        d["Log2FC"] = np.log2(d["TPM.%s" % ct2] / d["TPM.%s" % ct1])
        
        cutoff = np.log2(4)
        d1 = d[(d["TPM.Max"] >= 50) & (d["Log2FC"] >= cutoff)]
        d2 = d[(d["TPM.Max"] >= 50) & (d["Log2FC"] < -cutoff)]
        print("Up-regulated:", len(d1))
        print("Down-regulated:", len(d2))
        
        xs, ys = d["Log2FC"], np.log10(d["TPM.Max"] + 1)
        xs1, ys1 = d1["Log2FC"], np.log10(d1["TPM.Max"] + 1)
        xs2, ys2 = d2["Log2FC"], np.log10(d2["TPM.Max"] + 1)

        plt.figure(figsize=(3, 3))
        plt.scatter(xs, ys, marker=".", color="C7")
        plt.scatter(xs1, ys1, marker=".", color="C3")
        plt.scatter(xs2, ys2, marker=".", color="C0")

        plt.axvline(0, lw=1, ls="--", color="black")
        # plt.axhline(2, lw=1, ls="--", color="black")
        plt.xlim(-8, 8)
        plt.ylim(0, 5)
        plt.tight_layout()
        plt.show()
```
