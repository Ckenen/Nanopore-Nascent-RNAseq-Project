# Relationship between kinetic parameters

```python
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = pd.read_csv("../../1_FLAIRseq/reports/gene_based/K562.treatment_cells.min_read_2_min_tc_2.annotated.tsv", sep="\t", index_col=0)
```

```python
dat = dat[dat["GeneType"] == "protein_coding"]
dat = dat[dat["Total"] >= 50]
dat = dat[dat["Total"] > dat["Nascent"]]
dat = dat[dat["Nascent"] > 0]
dat["D"] = np.log(2) / dat["T"]
dat["K"] = dat["D"] * dat["TPM"] / 100
len(dat)
```

```python
xs = np.log2(dat["TPM"])
ys = np.log2(dat["K"])
print(spearmanr(xs, ys))
plt.figure(figsize=(3, 3))
sns.scatterplot(x=xs, y=ys, marker=".", color="black")
sns.kdeplot(x=xs, y=ys, fill=True, cmap="bwr")
plt.tight_layout()
plt.show()
```

```python
xs = np.log2(dat["TPM"])
ys = np.log2(dat["T"])
print(spearmanr(xs, ys))
plt.figure(figsize=(3, 3))
sns.scatterplot(x=xs, y=ys, marker=".", color="black")
sns.kdeplot(x=xs, y=ys, fill=True, cmap="bwr")
plt.tight_layout()
plt.show()
```

```python
xs = np.log2(dat["K"])
ys = np.log2(dat["T"])
print(spearmanr(xs, ys))
plt.figure(figsize=(3, 3))
sns.scatterplot(x=xs, y=ys, marker=".", color="black")
sns.kdeplot(x=xs, y=ys, fill=True, cmap="bwr")
plt.tight_layout()
plt.show()
```

```python
xs = np.log2(dat["K"])
ys = np.log2(dat["D"])
print(spearmanr(xs, ys))
plt.figure(figsize=(3, 3))
sns.scatterplot(x=xs, y=ys, marker=".", color="black")
sns.kdeplot(x=xs, y=ys, fill=True, cmap="bwr")
plt.tight_layout()
plt.show()
```
